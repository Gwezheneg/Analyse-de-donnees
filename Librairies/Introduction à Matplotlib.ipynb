## Création de Graphiques de Base
Pour réaliser un graphique avec la fonction pyplot, nous commençons par créer une figure en définissant sa zone, puis nous ajoutons et / ou modifions ses éléments à l’aide de fonctions offertes par pyplot.

Pour illustrer les fonctionnalités de matplotlib, nous allons avoir besoin de générer des valeurs, à l’aide de la librairie numpy, que nous chargeons :

````python
import numpy as np

# Données
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Création du graphique
plt.figure()             # Nous créons une figure vide (nous pouvons préciser ses dimensions)
plt.plot(x, y)           # Nous fournissons les coordonnées en abscisses et en ordonnées en premiers paramètres
plt.plot(x, y-1)        # Mettre une deuxième courbe sur le graphique

# Ajout de titres et de labels
plt.title("Graphique 1")
plt.xlabel("Axe X")
plt.ylabel("Axe Y")

# Ajout d'une grille de fond
plt.grid()

# Enregistrement de la figure dans le répertoire de travail
# plt.savefig("graphique_1.png")

# Affichage
plt.show()
````

### Paramètres esthétiques dans plot()

```
plt.plot(x, y, color='c', label='Sinusoïdal', linestyle='-', linewidth='2', marker='+',...)
```

- Alpha : transparence du graphique

- Color (c) : couleur de la ligne

- Label : utilisé avec le plt.legend()

- Linestyle (ls) : style de la ligne
    - '-' ligne pleine, 
    - '--' tiret, 
    -  '-.' tiret point, 
    - ':' pointillés, ...

- Linewidth (lw) : épaisseur du Linestyle

- Marker : 
    - 'o' cercle, 
    - '\*' étoile, 
    - '.' point, 
    - ','	Pixel, 
    - 'x' X, 
    - '+'	plus,
    - 's'	carré,
    - 'D'	diamant,
    - 'p'	pentagone,
    - 'h'	hexagone,
    - 'v'	triangle vers le bas,
    - '^'	triangle vers le haut,
    - '<'	triangle vers la gauche,
    - '>'	triangle vers la droite, ...

- Markeredgecolor (mec) : couleur du bord du marqueur

- Markeredgewidth (mew) : taille du bord du marqueur

- Markerfacecolor (mfc) : couleur du centre du marqueur

- Markersize (ms) : taille du centre du marqueur

## Paramètres de grid()

plt.grid() permet d'ajouter une grille de fond dans le graphique.

Nous pouvons utiliser un paramètre **axis** dans la fonction grid() pour spécifier le sens de la grille. Les valeurs possibles sont 'x', 'y' et 'both', par défaut la valeur est 'both'.

Nous pouvons également déterminer d'autres paramètres dont : grid(**color** = 'color', **linestyle** = 'linestyle', **linewidth** = number)

### Subplots
Avec subplot() nous pouvons créer plusieurs graphiques dans une même figure.

La syntaxe est : **plt.subplot(ligne, colonne, position)**


Le cycle de vie d'une figure multi-graphique est : 

````python
# Données
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Création d'une figure
plt.figure()

# Création du graphique 1
plt.subplot(2,1,1)
plt.plot(x, y1, c="r")
plt.title("Sin(x)")

# Création du graphique 2
plt.subplot(2,1,2)
plt.plot(x, y2, c="g")
plt.title("Cos(x)")

# Affichage de la figure
plt.show()
````

### Méthode orientée objet

````python
# Création des subplots
fig, axs = plt.subplots(2, 2, sharex=True) # sharex/sharey : les graphiques partagent le même axe des abscisses/ordonnées

# Premier subplot
axs[0][0].plot(x, y1)
axs[0][0].set_title('sin(x)')

# Deuxième subplot
axs[0][1].plot(x, y1-x**2, c='y')
axs[0][1].set_title('sin(x)-x**2')

# Troisième subplot
axs[1][0].plot(x, y2, color='red')
axs[1][0].set_title('cos(x)')

# Quatrième subplot
axs[1][1].plot(x, y2-x**2, color='g')
axs[1][1].set_title('cos(x)-x**2')

# Ajustement des espaces entre les subplots
plt.tight_layout()

# Titre pour la figure entière
plt.suptitle("Quelques fonctions")

# Affichage
plt.show()
````

**fig** est un objet et **ax** est un tableau ndarray qui contient des objets

## Histogramme

````python
# Données
data = np.random.randn(1000)

# Création de l'histogramme
plt.hist(data, bins=30, edgecolor='black') # bin : épaisseur des batons

# Ajout de titres et de labels
plt.title("Histogramme")
plt.xlabel("Valeur")
plt.ylabel("Fréquence")

# Affichage
plt.show()
````

### Diagramme à Barres

````python
# Données
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 40]

# Création du diagramme à barres
plt.bar(categories, values, color='blue', width = 0.5)

# Pour créer un diagramme à barres horizontales : plt.barh()

# Ajout de titres et de labels
plt.title("Diagramme à Barres")
plt.xlabel("Catégories")
plt.ylabel("Valeurs")

# Affichage
plt.show()
````

````python
# Série de bâtons côte-à-côte

# Données
pays = ["France", "Italie", "Belgique", "Allemagne"]
chomage_f = [9.1, 11.2, 6.4, 2.9]
chomage_h = [9.5, 9, 6.6, 3.8]

# Position sur l'axe des x pour chaque étiquette
position = np.arange(len(pays))

# Largeur des barres
largeur = .35

# Création de la figure et d'un set de sous-graphiques
fig, ax = plt.subplots()
r1 = ax.bar(position - largeur/2, chomage_f, largeur)
r2 = ax.bar(position + largeur/2, chomage_h, largeur)
plt.title('Chômage selon le sexe et le pays')

# Modification des marques sur l'axe des x et de leurs étiquettes
ax.set_xticks(position)
ax.set_xticklabels(pays)

# Affichage
plt.show()
````

````python
# Diagrammes en bâtons empilés

# Données
pays = ["France", "Italie", "Belgique", "Allemagne"]
nb_chomeurs_f = [1.307, 1.185, .577, .148]
nb_chomeurs_h = [1.46, 1.338, .878, .179]


plt.bar(pays, nb_chomeurs_f)
plt.bar(pays, nb_chomeurs_h, bottom = nb_chomeurs_f)

# Affichage
plt.show()
````

### Diagramme en Secteurs

````python
# Données
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
myexplode = [0.2, 0, 0, 0]

# Création du diagramme en secteurs
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode = myexplode)
plt.legend(title = "Test")

# Ajout de titre
plt.title("Diagramme en Secteurs")

# Affichage
plt.axis('equal')  # Assure que le diagramme est dessiné en cercle.
plt.show()
````

### Les paramètres de la fonction pie()

- labels : étiquette du graphique
- colors : couleurs des parts
- startangle : angle du début du graphique
- explode : espacement entre les parts
- shadows : ombre sous les parts

## Boxplot

````python
# Données
x = np.random.randn(1000)

# Création du graphique
plt.boxplot(x)

# Affichage
plt.show()


# En renseigant à False la valeur du paramètre vert, le boxplot est tracé horizontalement :

plt.boxplot(x, vert = False)
````

## ContourPlot()

````python
# Données
f = lambda x, y: np.sin(x) + np.cos(x-y)*np.cos(x)

X = np.linspace(0, 5, 150)
Y = np.linspace(0, 5, 150)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)

# Création du graphique
plt.contour(X, Y, Z, 40, colors='grey')

# Affichage
plt.show()
````

````python
# Le même graphique est une carte des couleurs
plt.contourf(X, Y, Z, 40, cmap='viridis')
plt.colorbar()

# Affichage
plt.show()
````

## Graphiques 3D

````python
from mpl_toolkits.mplot3d import Axes3D

# Données
f = lambda x, y: np.sin(x) + np.cos(x+y)

X = np.linspace(0, 5, 50)
Y = np.linspace(0, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)

# Création du graphique
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')

# Affichage
plt.show()
````

## Imshow()

Pour afficher des images, nous utilisons imshow()

````python
# Histogramme d'une image
from scipy import misc
face = misc.face(gray=True) # Nous mettons l'image en nuance de gris

# Création d'une figure
plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.imshow(face, cmap='gray') 
plt.subplot(122)
plt.hist(face.ravel(), bins=255) # Nous voulons voir la balance des blancs de l'image

# Affichage
plt.show()
````
