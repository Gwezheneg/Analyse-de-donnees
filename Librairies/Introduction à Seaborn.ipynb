## Importation

Pour utiliser Seaborn dans votre code Python, importez-le de la manière suivante :

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Application d'un thème esthétique
sns.set_theme()
```

## Jeux de Données Intégrés

Seaborn propose plusieurs jeux de données intégrés que vous pouvez utiliser pour pratiquer et tester vos visualisations.


```python
# Liste des jeux de données intégrés
print(sns.get_dataset_names())
```

```python
# Chargement du jeu de données 'tips'
data = sns.load_dataset('tips')

data.head()
```

## Visualisations de Base

### Graphique de Distribution (Histogramme et KDE)

```python
# Histogramme
sns.histplot(data['total_bill'], kde=False, bins=30)
plt.title('Répartition des factures')
plt.xlabel("Total Facture")
plt.ylabel("Fréquence")
```

### Diagramme en Barres

```python
# Diagramme en barres
sns.barplot(x='day', y='total_bill', data=data)
plt.title('Total des factures par jour')
plt.show()
```

### Catplot

```python
# Catplot
plt.figure(figsize=(32, 8))
sns.catplot(x='day', y='total_bill', hue='sex', data=data, kind='bar', height=6, aspect=1.5, errorbar='sd', col='smoker')
plt.xlabel('Jour de la semaine')
plt.ylabel('Total Facture')
plt.legend(title='Genre')

# Kind : bar, violin, box...
```

### Boxplot

```python
# Boxplot
sns.boxplot(x='day', y='total_bill', data=data, hue='sex')
plt.title('Boxplot de total_bill par jour')
plt.legend(title="Genre")
plt.show()
```

### Violinplot

```python
# Violinplot
sns.violinplot(x='day', y='total_bill', data=data, hue='sex', split=True, inner='quartile')
plt.title('Violinplot de total_bill par jour')
plt.legend(title="Genre")
plt.show()
```

### Scatterplot

```python
# Scatterplot
sns.scatterplot(x='total_bill', y='tip', data=data, hue="time", size="size")
plt.title('Scatterplot de total_bill vs tip')
plt.show()
```

### Stripplot

```python
# Stripplot
sns.stripplot(x='day', y='total_bill', data=data, hue='sex', jitter=True,  dodge=True)
```

## Visualisations Avancées

### Pairplot

```python
# Pairplot
sns.pairplot(data, height=5)
plt.show()
```

### PairGrid

```python
# PairGrid
grid = sns.PairGrid(data, hue='sex', height=5)
grid.map_lower(sns.kdeplot)
grid.map_upper(sns.scatterplot)
```

### Jointplot

```python
# Jointplot
sns.jointplot(x='total_bill', y='tip', data=data, kind='reg')
```

```python
# Jointplot
sns.jointplot(x='total_bill', y='tip', data=data, kind='kde')
```

### FacetGrid

```python
# FacetGrid
g = sns.FacetGrid(data, col='sex', row='time', height=10)
g.map(sns.scatterplot, 'total_bill', 'tip')
plt.show()
```

### Table de Pivot

```python
# Table de Pivot
table = data.pivot_table(index='sex', columns="smoker", aggfunc='size', observed=False)
table
```

### Heatmap

```python
# Heatmap
sns.heatmap(table, annot=True, cmap='coolwarm')
```

## Personnalisation des Graphiques

### Style et Contexte

```python
# Style
sns.set_style('whitegrid')

# Contexte
sns.set_context('talk')

# Création du graphique
sns.histplot(data['total_bill'], kde=True)
plt.title('Histogramme de total_bill avec style et contexte personnalisés', fontsize=12)
plt.show()
```

### Palette de Couleurs

```python
# Palette de couleurs
sns.set_palette('pastel')

# Création du graphique
sns.barplot(x='day', y='total_bill', data=data)
plt.title('Diagramme en barres avec palette de couleurs personnalisée')
plt.show()
```

## Enregistrement des Graphiques

```python
# Création du graphique
sns.boxplot(x='day', y='total_bill', data=data)
plt.title('Boxplot de total_bill par jour')

# plt.savefig('boxplot.png')

# Affichage
plt.show()
```
